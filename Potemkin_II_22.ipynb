Одномерный кристалл (30) 
Рассмотрите одномерный кристалл с двумя атомами различной массы $m$
и $M$ в элементарной ячейке, состоящий из $N$ элементарных ячеек (всего $2N$ атомов), замкнутых в кольцо
(периодические граничные условия).

Запишем Лагранжиан: $L = T - U = \frac{m}{2} \sum_{i=1}^{N} v_1^{2} $. Возьмем $R=1$, угол $\varphi_i$, 

Тогда $v_1=R \dot{\varphi_1}$ где $\dot{\varphi_1} = \frac{d\varphi_1}{dt}$.

Отсюда, $$
  L = T - U = \cfrac{m}{2} \sum_{i=1}^{N} \dot\varphi_{2i-1}^{2} + \cfrac{M}{2} \sum_{i=1}^{N} \dot\varphi_{2i}^{2} - \frac{k}{2}\sum_{i=1}^{2N-1} (\varphi_i - \varphi_{i+1} ) $$

$$\frac{d}{dt}\frac{dL}{d\dot\varphi_S}-\frac{dL}{d\varphi_S}=0,$$ 
Возьмем $S$ - нечетное.

Уравнение Ньютона: 
$$ \frac{d}{dt} m \dot\varphi_S + \frac{k}{2} \left[{ 2 \left( \varphi_S- \varphi_{S-1} \right) + 2 \left( \varphi_S + \varphi_{S+1} \right) }\right] = 0 $$
$$ m \ddot{\varphi} = k\left[2\varphi_S - \varphi_{S-1} - \varphi_{S+1} \right] $$

$S$ - четное:

$$M \ddot{\varphi_S} = k\left(2\varphi_S - \varphi_{S-1} - \varphi_{S+1} \right) $$

$2N$ степеней свободы, $2N-1$ должно быть
Отсутствие вращение как целого: $\vec{M}=0$ - момент импульса, $\Rightarrow$ $m \sum \varphi_{2i-1} + M \sum \varphi_{2i}=const=0$

Выписываем уравнения: 
$$
\sqrt{1}m:\ddot\varphi_{1}=\frac{k}{m}[2\varphi_1 - \varphi_{2N} -\varphi_2], \\
 \sqrt{2}M:\ddot\varphi_{2}=\frac{k}{M}[2\varphi_2 - \varphi_{1} -\varphi_3], \\
 \vdots  \\
\frac{m}{\sqrt{2N-1}}: \ddot\varphi_{2N-1}=\frac{k}{m}[2\varphi_{2N-1} - \varphi_{2N-2} -\varphi_{2N}]
$$
или 
$$
 \begin{cases}
   \ddot\varphi_{1}=\frac{k}{m}[2\varphi_1 - \varphi_{2N} -\varphi_2]
   \\
 \ddot\varphi_{2}=\frac{k}{M}[2\varphi_2 - \varphi_{1} -\varphi_3] 
 \\
 \vdots  
 \\
\ddot\varphi_{2N-1}=\frac{k}{m}[2\varphi_{2N-1} - \varphi_{2N-2} -\varphi_{2N}]
\end{cases}
$$


2) Предполагая, что все атомы движутся с одной и той же частотой, $x_i(t) = u_i e^{−iωt}$,
перепишите найденые выше уравнения в виде системы линейных уравнений на вектор $u$. Составьте матрицу $A$, 
спектр которой определяет частоты нормальных мод.

Запишем в векторно - матричном виде:
$$
\begin{equation}
\frac{d^2}{dt^2} 
    \begin{pmatrix}
    \varphi_1 \\ 
    \vdots \\ 
    \varphi_{2N-1}
    \end{pmatrix}
= \frac{k}{m} A 
    \begin{pmatrix}
    \varphi_1 \\ 
    \vdots \\ 
    \varphi_{2N-1}
    \end{pmatrix}
\end{equation}
$$
Распишем:
$$
    m(\varphi_1+\varphi_3+\cdots+\varphi_{2N-1})+M(\varphi_2+\cdots+\varphi_{2N})=0 \\
    \varphi_{2N}=-(\varphi_2+\cdots+\varphi_{2N-2})-\frac{m}{M}(\varphi_1+\cdots+\varphi_{N-1}) \\
    \Rightarrow \ddot\varphi_{1}=\frac{k}{m}[2\varphi_1 - \varphi_{2N} -\varphi_2] = \frac{k}{m} [2\varphi_2+\varphi_4+\cdots+\varphi_{2N-2}+\frac{m}{M}(\varphi_1+\cdots+\varphi_{2N-1})] \\
    \Rightarrow \ddot\varphi_{2N-1}=\frac{k}{m}[2\varphi_{2N-1} - \varphi_{2N-2} -\varphi_{2N}] = \frac{k}{m} \left[ \varphi_{2N-1}+\varphi_2+\cdots+\frac{m}{M}(\varphi_{1}+\cdots+\varphi_{2N-1}) \right]
$$
Тогда, матрица А:
$$ 
A = 
\begin{pmatrix}
2+\frac{m}{M} & 0 & \frac{m}{M} & 1 & \frac{m}{M} & 1\\ 
-\frac{m}{M} & 2\frac{m}{M} & -\frac{m}{M} & \ddots & \ddots & \vdots \\ 
\vdots & -1 & 2 & -1 & \ddots & \vdots\\ 
\vdots & \ddots & -\frac{m}{M} & 2\frac{m}{M} & -\frac{m}{M} & \vdots \\
\frac{m}{M} & 1 & \cdots & \cdots & \cdots & 2+\frac{m}{M}
\end{pmatrix}
$$

"Расщепим" систему, и переименуем: $V^T \ddot{\vec\varphi} = \ddot{\vec{y}}$, $V^T\vec\varphi = \vec{y}$
$$
\ddot{\vec\varphi}=\frac{k}{m} A \vec\varphi; \quad V^T \ddot{\vec\varphi} = \frac{k}{m}V^TAVV^T\vec\varphi \quad \Rightarrow \ddot{\vec{y}}=-\frac{k}{m} \Lambda\vec{y}
$$
Перешли к независимым уравнениям. $\Lambda$ - диагональная, по диагонали $(\lambda_1, \cdots, \lambda_{2N-1})$.
$$ \Rightarrow \ddot{y_S} = -\frac{k}{m}\lambda_S y_S $$
Получили уравнение вида $$\ddot{y_S}+\omega_S^2y_S=0, \ \ S=1\cdots 2N-1$$
решение которое в виде $$ y_S = \textrm{sin}\omega_S t + \textrm{cos}\omega_S t, \quad \omega_S = \sqrt{\frac{k}{m}\lambda_S}$$
$y_S$ - нормальные моды, $w_S$ - частоты нормальных мод

3) Используя $np.linalg.eig$, найдите спектр матрицы $A$ (возьмите $N = 1000$ и $M/m = 2$). Постройте
гистограмму собственных значений. Обратите внимание, что в спектре есть щель – ‘запрещенная’
область энергии внутри спектра, которая разделяет ‘разрешенную’ область на две части.

```python
N=1000
k=1/2
m=1000
A=np.zeros([2*N-1,2*N-1])
A[0,0]=2+k
A[0,2:-1]=[k, 1]*(N-2)
A[0,-1]=k
for i in range(1,2*N-1,2):
    A[i, i-1:i+2]=[-k, 2*k, -k]
for i in range(2,2*N-2,2):
    A[i, i-1:i+2]=[-1., 2., -1.]
A[-1,:-3]=[k, 1]*(N-2)
A[-1,-3]=k
A[-1,-1]=2+k
l, V = np.linalg.eig(A)
plt.hist(l.real);
```
